```python
import matplotlib.pyplot as plt
import networkx as nx

#green line
MyGraph = nx.Graph()
MyGraph.add_node('A', npos=(10, 10), ccn='#0000FF')
MyGraph.add_node('B', npos=(20, 25), ccn='#0000FF')
MyGraph.add_node('C', npos=(47, 25), ccn='#0000FF')
MyGraph.add_node('D', npos=(70, 25), ccn='#0000FF')
MyGraph.add_node('E', npos=(85, 35), ccn='#0000FF')
MyGraph.add_node('F', npos=(105, 50), ccn='#0000FF')

#Connection for the greenline
MyGraph.add_edge('A', 'B', cce='#0000FF')
MyGraph.add_edge('B', 'C', cce='#0000FF')
MyGraph.add_edge('C', 'D', cce='#0000FF')
MyGraph.add_edge('D', 'E', cce='#0000FF')
MyGraph.add_edge('E', 'F', cce='#0000FF')


#Extract attributes from graph to dictionary
pos = nx.get_node_attributes(MyGraph,'npos')
nodecolour = nx.get_node_attributes(MyGraph, 'ccn')
edgecolour = nx.get_edge_attributes(MyGraph, 'cce')

#Place dictionary in lists
NodeList = list(nodecolour.values())
EdgeList= list(edgecolour.values()) 

#Display the station name
plt.text(13, 10, s='Hyde Park Corner', rotation=0)
plt.text(13, 28, s='Green Park', rotation=0)
plt.text(45, 5, s='Piccadilly Circus', rotation=-90)
plt.text(75, 24, s='Leicester Square', rotation=0)
plt.text(55, 35, s='Convent Garden', rotation=0)
plt.text(87, 50, s='Holborn', rotation=0)

#Display the Miles/Kilometer
plt.text(15, 15, s='0.7 Miles')
plt.text(25, 22, s='0.4 Miles')
plt.text(52, 22, s='0.1 Miles')
plt.text(77, 28, s='0.6 Miles')
plt.text(95, 40, s='1.2 Miles')

#Display explanation box
plt.text(105, 3, "Key", fontsize=12, fontweight='bold')
plt.plot(6.5, -1.5, color='blue', label='Piccadilly Line')
plt.legend(loc='lower right')

#Draw Graph
nx.draw_networkx(MyGraph, pos, node_color=NodeList)
nx.draw_networkx_edges(MyGraph, pos, edge_color=EdgeList)

#Show graph
plt.show()
```
